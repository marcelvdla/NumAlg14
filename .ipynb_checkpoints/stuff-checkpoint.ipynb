```python
import numpy as np

def x2(x):
    return x ** 2

def _2x(x):
    return 2*x

def condition(x):
    return x/np.tan(x)
```

```python
x = 2
dx = 0.01

print(x2(x + dx) - x2(x), _2x(x)*dx)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(1,10, 10**4)
y = condition(x)

fig, ax = plt.subplots()
plt.plot(x,y)

ax.set_ylim(-50, 50)
plt.show()
```
